# Demo on Fourier Transform and Truncation

### 任务：读取 NCEP/NCAR Reanalysis 2016年1月21日沿55N度纬圈的瞬时 500mb 高度场，通过傅立叶正变换和反变换，熟悉傅立叶变换的概念和波数截断的概念，并体会谱变换中收敛速度的快慢

## 1. 读入数据

```python
%matplotlib inline

import xarray as xr
import numpy as np
import pkuplot

# Open netCDF file
f = xr.open_dataset('hgt.2016.nc',decode_times=False)

lat = f.lat
lon = f.lon

Nlat = len(lat)
Nlon = len(lon)

print("N_lat =",Nlat)
print("N_lon =",Nlon)

# Read data
z500 = f.hgt.sel(lat=55,level=500).isel(time=20)  # 2016 Jan-21
```

## 2. Fourier Analysis (transform, DFT)

```python
C = np.fft.rfft(z500)
print("原始数据 z500 shape =",z500.shape)
print("复谱系数 C    shape =",C.shape)

C_amp = np.abs(C)/Nlon
pkuplot.plot2d(1,[range(len(C_amp)),],[C_amp,],style=['b-o',],lw=[4,],lb=['2016-Jan-21 at 55N',],
        ti='Spectrum of Z500 Waves',xl='Wave Number',yl='Amplitude (gpm)',xlim=[0,10],ylim=[0,100])

for i in range(11): print('Wave #',i,':   ',C_amp[i])
```

## 3. Fourier Synthesis (inverse transform, iDFT)

```python
def tr_syn(C,keep_to_wavenumber):
    '''
    Truncate off, then Synthesis back
    '''
    Ct = C.copy()
    Ct[keep_to_wavenumber+1:] = 0
    return np.fft.irfft(Ct)

z500_0 = tr_syn(C,0)
z500_1 = tr_syn(C,1)
z500_2 = tr_syn(C,2)
z500_3 = tr_syn(C,3)
z500_7 = tr_syn(C,7)

# plot
pkuplot.plot2d(6,[lon,lon,lon,lon,lon,lon],[z500,z500_0,z500_1,z500_2,z500_3,z500_7],
               style=['c-','y--','k-','r-','b-','g--'],
               lw=[10,2,2,2,2,2],
               lb=['raw','just W#0','to W#1','to W#2','to W#3','to W#7'],
               ti='2016-Jan-21',xl='Lon',yl='Z500 (gpm)',xlim=[-30,390])
```

## 4. Check each wave number, in terms of AMPLITUDE and PHASE

```python
def wavenumber(C,wn):
    '''
    Only keep one single wavenumber, then Synthesis back
    '''
    Ct = C.copy()*0
    Ct[wn] = C[wn]

    return np.fft.irfft(Ct)

z500_0 = wavenumber(C,0)
z500_1 = wavenumber(C,1)
z500_2 = wavenumber(C,2)
z500_3 = wavenumber(C,3)
z500_7 = wavenumber(C,7)

# plot
pkuplot.plot2d(4,[lon,lon,lon,lon],[z500_1,z500_2,z500_3,z500_7],
               style=['k-','r-','b-','g--'],
               lw=[2,2,2,2],
               lb=['just W#1','just W#2','just W#3','just W#7'],
               ti='2016-Jan-21',xl='Lon',yl='Z500 (gpm)',xlim=[-30,390])
```
