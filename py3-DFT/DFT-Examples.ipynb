# Fourier Transform 究竟有什么用？

## 1. 沿着某一个纬度的气象变量 (1D-DFT)

### 读入必要的库

```python
%matplotlib inline

import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import imread, imsave
import pkuplot
```

### 打开气候态（1981-2010）和2000年的高度场文件

```python
f_ltm = xr.open_dataset('hgt.mon.1981-2010.ltm.nc',decode_times=False)
f_day = xr.open_dataset('hgt.2000.nc',decode_times=False)

lat = f_ltm.lat
lon = f_ltm.lon

Nlat = len(lat)
Nlon = len(lon)

print("N_lat =",Nlat)
print("N_lon =",Nlon)
```

### 读入1月北纬55度（北半球，冬天）的Z500

```python
# Read data
n1_ltm = f_ltm.hgt.sel(lat=55,level=500).isel(time=0)   # Climatological Jan
n1_day = f_day.hgt.sel(lat=55,level=500).isel(time=15)  #           2000 Jan-15

# Quick view
pkuplot.plot2d(2,[lon,lon],[n1_ltm,n1_day],
               style=['r-','b-'],lw=[2,2],lb=['Jan, Climatology','Jan-15, 2000'],
               ti='Space Domain: NH 55N',xl='Lon',yl='Z500 (gpm)',xlim=[-10,370])

# Fourier analysis
Cn1_ltm = np.fft.rfft(n1_ltm)
Cn1_ltm_abs = np.abs(Cn1_ltm)/Nlon

Cn1_day = np.fft.rfft(n1_day)
Cn1_day_abs = np.abs(Cn1_day)/Nlon

xtmp = range(len(Cn1_ltm_abs))

pkuplot.plot2d(2,[xtmp,xtmp],[Cn1_ltm_abs,Cn1_day_abs],
               style=['r-o','b-s'],lw=[4,4],lb=['Jan, Climatology','Jan-15, 2000'],
               ti='Spectral Domain: NH 55N',xl='Wave #',yl='Amplitude',
               xlim=[0,20],ylim=[0,150],ylog=False)

# Print
for i in range(11):
    print("wave #",i,"Red=",Cn1_ltm_abs[i],"Blue=",Cn1_day_abs[i])
```

### 读入1月南纬55度（南半球，夏天）的Z500

```python
# Read data
s1_ltm = f_ltm.hgt.sel(lat=-55,level=500).isel(time=0)
s1_day = f_day.hgt.sel(lat=-55,level=500).isel(time=15)

# Quick view
pkuplot.plot2d(2,[lon,lon],[s1_ltm,s1_day],
               style=['r-','b-'],lw=[2,2],lb=['Jan, Climatology','Jan-15, 2000'],
               ti='Space Domain: SH 55S',xl='Lon',yl='Z500 (gpm)',xlim=[-10,370])

# Fourier analysis
Cs1_ltm = np.fft.rfft(s1_ltm)
Cs1_ltm_abs = np.abs(Cs1_ltm)/Nlon

Cs1_day = np.fft.rfft(s1_day)
Cs1_day_abs = np.abs(Cs1_day)/Nlon

xtmp = range(len(Cs1_ltm_abs))

pkuplot.plot2d(2,[xtmp,xtmp],[Cs1_ltm_abs,Cs1_day_abs],
               style=['r-o','b-s'],lw=[4,4],lb=['Jan, Climatology','Jan-15, 2000'],
               ti='Spectral Domain: SH 55N',xl='Wave #',yl='Amplitude',
               xlim=[0,20],ylim=[0,150],ylog=False)

# Print
for i in range(11):
    print("wave #",i,"Red=",Cs1_ltm_abs[i],"Blue=",Cs1_day_abs[i])
```

## 2. 二维数据的Fourier变换 (2D-DFT)

### 读入2000-Jan-15的Z500全场数据

```python
z500 = f_day.hgt.sel(level=500).isel(time=15)
z500.plot()                                       # Looks like this 原始数据是酱婶儿的
```

### 对这个Z500做2D FFT变换

#### 注意：这个变换的先验假设并不完美。主要原因是，数据的东西方向是物理上的cyclic，但数据的南北方向并不是物理上的cyclic。这使得南北边界处的差异可能过大，会保留在高频部分的谱系数里。

```python
Cz500 = np.fft.rfft2(z500)

# 2D 振幅
Cz500_abs = xr.DataArray(np.abs(Cz500))
Cz500_abs.plot()
```

```python
# 2D 位相
Cz500_ang = xr.DataArray(np.angle(Cz500))
Cz500_ang.plot()
```

### 反变换是完全没有问题滴

```python
z500new = xr.DataArray(np.fft.irfft2(Cz500))
z500new.plot()
```

### 如果对谱系数做截断，则返回格点空间后会显得“模糊”，但大尺度特征已被很好滴保留，“抓大放小”

```python
Ktrunc = 41

# Method 1
Cz500_trunc = Cz500.copy()
Cz500_trunc[:,Ktrunc:] = 0

# iDFT
z500_trunc = xr.DataArray(np.fft.irfft2(Cz500_trunc))
z500_trunc.plot()

print("Z500 的维度是：",z500.shape)
print("其谱系数的维度是：",Cz500.shape)
print("截断后的维度是：",Cz500[:,:Ktrunc].shape)
```

## 3. 照片也是数据，也可以进行 Fourier Analysis (2D-DFT)

### 比如读入我们的全班合影照片（宽1600点，高947点，RGB三色）

```python
pic = imread('./cm2018portrait-2018mar12.jpg').astype(np.float32)
pic = pic/255
print(pic.shape)

plt.imshow(pic)
```

### 把照片按 RGB 三色分别进行Fourier变换、截断、再反变换，再三色合成为新图片

```python
# RGB 颜色分解
c_red   = pic[:,:,0]
c_green = pic[:,:,1]
c_blue  = pic[:,:,2]

# DFT
C_c_red   = np.fft.rfft2(c_red)
C_c_green = np.fft.rfft2(c_green)
C_c_blue  = np.fft.rfft2(c_blue)

# Truncation
Ktrunc = 1000
C_c_red[Ktrunc:,Ktrunc:]   = 0      # 第一个维度也许不该截断
C_c_green[Ktrunc:,Ktrunc:] = 0
C_c_blue[Ktrunc:,Ktrunc:]  = 0

# iDFT
pic_new = pic.copy()
pic_new[:,:,0] = np.fft.irfft2(C_c_red)
pic_new[:,:,1] = np.fft.irfft2(C_c_green)
pic_new[:,:,2] = np.fft.irfft2(C_c_blue)

plt.imshow(pic_new)
imsave('pic_truncated.jpg', pic_new)
```
